```python
#‡∏â‡∏±‡∏ô‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏´‡πâ‡πÄ‡πÄ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏Å‡πà‡∏≠‡∏ô‡πÄ‡πÄ‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏£‡∏≠‡∏á‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏ü‡∏•‡πå‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ù‡∏ô‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á
import pandas as pd
import os

# ===============================
# 1) Path ‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô
# ===============================
base_path = "/content/drive/MyDrive/data_midterm/Input - (Operation Daily rain from API_ 2025)"

month_folders = [
    "2025-07",
    "2025-08",
    "2025-09",
    "2025-10",
    "2025-11",
    "2025-12"
]

# ===============================
# 2) ‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏ü‡∏•‡πå‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ (Master)
# ===============================
# ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç: ‡πÉ‡∏ä‡πâ DataFrame ‡∏ó‡∏µ‡πà‡πÇ‡∏´‡∏•‡∏î‡πÑ‡∏ß‡πâ‡πÅ‡∏•‡πâ‡∏ß‡∏à‡∏≤‡∏Å dataframes dictionary ‡πÅ‡∏ó‡∏ô‡∏Å‡∏≤‡∏£‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏ü‡∏•‡πå‡∏ã‡πâ‡∏≥
stations_hii = dataframes['Stations_in_the_model_HII']
valid_station_nodes = set(stations_hii['Node'])

print(f"‚úÖ Stations in HII model: {len(valid_station_nodes)}")

# ===============================
# 3) ‡∏≠‡πà‡∏≤‡∏ô + ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏Å‡πà‡∏≠‡∏ô‚Äì‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏£‡∏≠‡∏á
# ===============================
filtered_monthly_dfs = {}

print("\n================ BEFORE vs AFTER FILTERING ================\n")

for month in month_folders:
    folder_path = os.path.join(base_path, month)
    csv_files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]

    raw_rows = 0
    raw_stations = set()

    filtered_rows = 0
    filtered_stations = set()

    for file in csv_files:
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)

        # ----- BEFORE FILTERING ----- (Only count if 'tele_station_oldcode' exists)
        if 'tele_station_oldcode' in df.columns:
            raw_rows += df.shape[0]
            raw_stations.update(df['tele_station_oldcode'].dropna().unique())

            # ----- AFTER FILTERING -----
            df_filtered = df[df['tele_station_oldcode'].isin(valid_station_nodes)]
            filtered_rows += df_filtered.shape[0]
            filtered_stations.update(df_filtered['tele_station_oldcode'].dropna().unique())
        else:
            print(f"   ‚ö†Ô∏è Warning: '{file}' in '{month}' does not contain 'tele_station_oldcode' column. Skipping this file for raw counts and filtering.")

    # ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏£‡∏≠‡∏á‡πÑ‡∏ß‡πâ‡πÉ‡∏ä‡πâ‡∏ï‡πà‡∏≠
    # ‡∏ï‡πâ‡∏≠‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á monthly_dfs_list ‡πÉ‡∏´‡∏°‡πà‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏™‡πà‡∏ß‡∏ô‡∏ô‡∏µ‡πâ ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏£‡∏ß‡∏°‡πÑ‡∏ü‡∏•‡πå‡πÑ‡∏î‡πâ‡πÅ‡∏°‡πâ‡∏à‡∏∞‡∏°‡∏µ warning
    monthly_dfs_list_for_concat = []
    for f in csv_files:
        file_path_current = os.path.join(folder_path, f)
        df_temp = pd.read_csv(file_path_current)
        if 'tele_station_oldcode' in df_temp.columns:
            monthly_dfs_list_for_concat.append(df_temp.query("tele_station_oldcode in @valid_station_nodes"))

    if monthly_dfs_list_for_concat:
        filtered_monthly_dfs[month] = pd.concat(monthly_dfs_list_for_concat, ignore_index=True)

    # ----- ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏• -----
    print(f"üìÖ Month: {month}")
    print(f"  BEFORE  ‚Üí Rows: {raw_rows:,} | Stations: {len(raw_stations)}")
    print(f"  AFTER   ‚Üí Rows: {filtered_rows:,} | Stations: {len(filtered_stations)}")
    print(f"  REMOVED ‚Üí Rows: {raw_rows - filtered_rows:,} | Stations: {len(raw_stations - filtered_stations)}")
    print("-" * 55)

# ===============================
# 4) ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏£‡∏≠‡∏á‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
# ===============================
if filtered_monthly_dfs:
    monthly_rain_filtered = pd.concat(filtered_monthly_dfs.values(), ignore_index=True)
    print("\n‚úÖ FINAL DATASET (After Filtering)")
    monthly_rain_filtered.info()
else:
    print("\n‚ùå No filtered dataframes were created. Check for errors in loading or filtering.")
```

Load the historical rainfall data (`hii_daily_rain`) and filter it to include only the `remaining_stations`. Convert relevant date columns to datetime objects and extract the 'Day of Year'. Ensure the 'data' column (rainfall value) is clean and ready for median calculation.

This addresses all instructions for the subtask.



```python
#‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏¥‡∏° median ‡∏ï‡∏≤‡∏° day of year ‡∏à‡∏≤‡∏Å Historical ‡πÅ‡∏¢‡∏Å‡∏ï‡∏≤‡∏°‡∏™‡∏ñ‡∏≤‡∏ô‡∏µ‡πÉ‡∏´‡πâ‡∏´‡∏ô‡πà‡∏≠‡∏¢

import pandas as pd

# 1. Extract the unique tele_station_oldcode values from the stations_remaining_df DataFrame
remaining_station_codes = stations_remaining_df['tele_station_oldcode'].unique()

print(f"Number of remaining stations after filtering: {len(remaining_station_codes)}")

# 2. Filter the hii_daily_rain DataFrame to include only the remaining_station_codes
historical_daily_rain = dataframes['hii_daily_rain']
historical_data_remaining_stations = historical_daily_rain[
    historical_daily_rain['station_code'].isin(remaining_station_codes)
].copy() # Use .copy() to avoid SettingWithCopyWarning

print(f"Shape of historical data for remaining stations: {historical_data_remaining_stations.shape}")

# 3. Convert the measure_datetime column to datetime objects
historical_data_remaining_stations['measure_datetime'] = pd.to_datetime(
    historical_data_remaining_stations['measure_datetime']
)

# 4. Extract the 'Day of Year' from the measure_datetime column
historical_data_remaining_stations['day_of_year'] = historical_data_remaining_stations['measure_datetime'].dt.dayofyear

print("First 5 rows of historical_data_remaining_stations with 'day_of_year':")
print(historical_data_remaining_stations.head())
```

**Reasoning**:
The next step is to calculate the historical median rainfall for each station and day of the year. This median will be used to impute missing rainfall values in the current data. Grouping by `station_code` and `day_of_year` and calculating the median of the `data` column directly fulfills this requirement.



```python
import pandas as pd

# 5. Calculate the historical median rainfall by station_code and day_of_year
median_historical_rainfall = historical_data_remaining_stations.groupby(
    ['station_code', 'day_of_year']
)['data'].median().reset_index()

# Rename the 'data' column to something more descriptive, like 'median_historical_rainfall'
median_historical_rainfall.rename(columns={'data': 'median_historical_rainfall'}, inplace=True)

print("First 5 rows of median_historical_rainfall:")
print(median_historical_rainfall.head())
print(f"Shape of median_historical_rainfall: {median_historical_rainfall.shape}")
```

**Reasoning**:
To prepare the current data for imputation, I need to filter `monthly_rain_filtered` to include only the `remaining_stations`, convert the `measure_datetime` column to datetime objects, and extract the 'Day of Year' to match the `median_historical_rainfall` DataFrame. This directly addresses the 'Prepare Current Data for Imputation' part of the task.



```python
import pandas as pd

# 6. Filter the monthly_rain_filtered DataFrame to include only the remaining_stations
current_data_for_imputation = monthly_rain_filtered[
    monthly_rain_filtered['tele_station_oldcode'].isin(remaining_station_codes)
].copy() # Use .copy() to avoid SettingWithCopyWarning

print(f"Shape of current data for imputation (remaining stations only): {current_data_for_imputation.shape}")

# 7. Convert the measure_datetime column to datetime objects in the current data
current_data_for_imputation['measure_datetime'] = pd.to_datetime(
    current_data_for_imputation['measure_datetime']
)

# 8. Extract the 'Day of Year' from measure_datetime in the current data
current_data_for_imputation['day_of_year'] = current_data_for_imputation['measure_datetime'].dt.dayofyear

# 9. Ensure the tele_station_oldcode column is consistent (it already is, but good to check/cast if needed)
current_data_for_imputation['tele_station_oldcode'] = current_data_for_imputation['tele_station_oldcode'].astype(str)

print("First 5 rows of current_data_for_imputation with 'day_of_year':")
print(current_data_for_imputation.head())
```

```python
import numpy as np

# 10. Merge current_data_for_imputation with median_historical_rainfall
# Ensure column names match for merging: 'tele_station_oldcode' for current data, 'station_code' for historical median
current_data_imputed = pd.merge(
    current_data_for_imputation,
    median_historical_rainfall,
    left_on=['tele_station_oldcode', 'day_of_year'],
    right_on=['station_code', 'day_of_year'],
    how='left'
)

# 11. Impute missing 'rainfall_value' with 'median_historical_rainfall'
# Create a mask for rows where rainfall_value is NaN
missing_rainfall_mask = current_data_imputed['rainfall_value'].isna()

# Fill NaN values in 'rainfall_value' using 'median_historical_rainfall'
current_data_imputed.loc[missing_rainfall_mask, 'rainfall_value'] = \
    current_data_imputed.loc[missing_rainfall_mask, 'median_historical_rainfall']

print(f"Shape of current_data_imputed after merge and imputation: {current_data_imputed.shape}")

# Verify if any NaN values remain in 'rainfall_value'
print(f"Number of remaining NaN values in 'rainfall_value': {current_data_imputed['rainfall_value'].isna().sum()}")

# Drop the 'station_code' (duplicate of tele_station_oldcode) and 'median_historical_rainfall' columns
current_data_imputed.drop(columns=['station_code', 'median_historical_rainfall'], inplace=True)

print("First 10 rows of current_data_imputed after imputation and cleanup:")
print(current_data_imputed.head(10))
```

```python
import pandas as pd

# 1. Extract year and month from the 'measure_datetime' column
current_data_imputed['year'] = current_data_imputed['measure_datetime'].dt.year
current_data_imputed['month'] = current_data_imputed['measure_datetime'].dt.month

# 2. Group by station, year, and month, and sum the rainfall_value
monthly_rainfall_calculated = current_data_imputed.groupby(
    ['tele_station_oldcode', 'year', 'month']
)['rainfall_value'].sum().reset_index()

# 3. Rename columns to match a more common format if desired (e.g., Node, rain)
monthly_rainfall_calculated.rename(columns={
    'tele_station_oldcode': 'Node',
    'rainfall_value': 'rain'
}, inplace=True)

# 4. Display the first few rows and info of the resulting DataFrame
print("--- Calculated Monthly Rainfall (First 5 rows) ---")
print(monthly_rainfall_calculated.head())

print("\n--- DataFrame Info for Calculated Monthly Rainfall ---")
monthly_rainfall_calculated.info()
```
